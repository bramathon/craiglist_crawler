```python
import numpy as np
import pandas as pd
import plotly
import plotly.express as px
import sklearn
from sklearn import metrics
from data_utils import load_dataset, core_metrics#, map_model
```

```python
X_train, X_test, y_train, y_test, X_scaler, y_scaler = load_dataset(scale=True,one_hot=True,categorize_bedrooms=True)
```

```python
from sklearn import neighbors
```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors=50)
knn.fit(X_train, y_train)
```

```python
y_pred = y_scaler.inverse_transform(knn.predict(X_test))
y_real = y_scaler.inverse_transform(y_test)
core_metrics(y_real,y_pred)
# df=pd.DataFrame({'error':(y_pred-y_real).reshape(-1)})
# px.histogram(df,x='error')
```
